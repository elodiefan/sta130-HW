**Hypothesis Testing, Null Hypothesis, and Alternative Hypothesis**

- The key factor differentiating testable and untestable ideas statistically is the ability to frame the idea as a statement about a population parameter that can be quantified and measured. For instance, you can test the idea "people are taller now than they were 50 years ago" because you can compare average heights of populations at different times. You cannot, however, test the idea "cats are funnier than dogs" because "funniness" is not something that can be objectively quantified and measured across populations.
- A good null hypothesis (often denoted as H0) should meet these criteria:
    - It's a statement about a population parameter.
    - It posits no effect or no difference.
    - It can be contradicted by evidence found in the data.
    - It's testable, meaning data can be collected and analyzed to assess its validity.
- In hypothesis testing, you contrast two competing hypotheses:
    - Null Hypothesis (* H0): This is the statement you're trying to disprove. It assumes the effect you're interested in doesn't exist in the population. For instance, if you were testing whether a new drug lowers blood pressure, the null hypothesis might be "the new drug has no effect on blood pressure."
    - Alternative Hypothesis (* HA): This is the statement you're trying to find evidence for. It proposes that there is an effect. In the drug example, the alternative hypothesis would be "the new drug does have an effect on blood pressure."

**Distinguishing Between Sample Statistics and Population Parameters in Hypothesis Testing**

**For a non-statistical audience:** 
- $x_i$: This represents a single observation or data point in a sample. For instance, if we are trying to measure the height of students in a class, each student's height would be an $x_i$.
- $\bar{x}$ (x-bar): This is the sample mean, calculated by averaging all the $x_i$ values in the sample. It gives an estimate of the central tendency of the sample data.
- $\mu$ (mu): This is the population mean, which is the true average value of the variable of interest for the entire population. This is usually unknown and it is what we are trying to learn about through hypothesis testing.
- $\mu_0$ (mu-zero): This is the hypothesized population mean, which is a specific value we are testing against in our hypothesis test. For example, we might test if the average height of students in a school ($\mu$) is equal to a hypothesized value, $\mu_0$ = 5'8".


- When we perform a hypothesis test, we are using the information from our sample (the $x_i$'s and $\bar{x}$) to make an inference about the population parameter, $\mu$. The result of the hypothesis test (whether we reject or fail to reject the null hypothesis) tells us something about the plausibility of $\mu_0$ being the true population mean, $\mu$. We are not directly testing the sample mean, $\bar{x}$ , it is simply a tool we use to reach a conclusion about the population.
- Overall speaking, hypothesis testing allows us to use limited sample data to make informed statements about the larger population from which the sample was drawn.

In more simple terms:


- The Big Picture (Population): If let's say you're interested in the average height of all students in a school. The "population" is the entire group of students, and the "population parameter" is the average height of everyone in this group, represented by  in statistical notation.
- The Snapshot (Sample): Since it is impractical to measure every student's height, you can take a smaller group of students, which is a "sample," and measure their heights. These individual measurements from your sample are your $x_i$'s.
- Estimating from the Snapshot (Sample Mean): You can calculate the average height of your sample, which is the "sample mean" represented by $\bar{x}$. This sample mean is a "statistic," meaning it is a value calculated from your sample data.
- Hypothesis Testing: Make an educated guess about the Big Picture: The goal of hypothesis testing is to use the information from your sample (the snapshot) to make an educated guess about the population (the big picture).
    - You start with an assumption about the population mean called the null hypothesis ($H_0$). For example, you might assume the average height of all students in the school is 5'8'' ($\mu_0$).
    - You then see if the data from your sample provides enough evidence to reject this null hypothesis. This doesn't mean proving the null hypothesis is definitively false; it means determining if your sample data is unlikely to occur if the null hypothesis were actually true for the entire population.
- The Result is About the Big Picture (Population): The outcome of your hypothesis test isn't telling you for sure whether the average height of your specific sample ($\bar{x}$) is equal to, greater than, or less than 5'8". Instead, it's giving you a level of confidence about whether the true average height of the entire school population () is likely to be 5'8", based on the evidence you gathered from your sample.

**Why Imagine a World Where the Null Hypothesis is True?**

The second "Pre-lecture" video explains that when calculating a p-value, we "imagine a world where the null hypothesis is true" because we are trying to determine how likely it is to observe our data, or data even more extreme, if the null hypothesis was actually true. 

*   This is where the **sampling distribution of the test statistic under the null hypothesis** becomes important.
*   The **sampling distribution** refers to the distribution of test statistics we would obtain if we were to repeatedly sample from the population under the assumption that the null hypothesis is true.
*   We can think about simulating this by, for example, flipping a coin many times if our null hypothesis is that the coin is fair. 

Essentially, the **p-value** represents how likely it would be to observe a test statistic as extreme as the one we actually calculated (or even more extreme), if we were to live in a world where the null hypothesis is true. 

* A small p-value tells us that it would be very unlikely to observe our data if the null hypothesis was true, suggesting the null hypothesis is likely false. Conversely, a large p-value doesn't necessarily prove the null hypothesis true, but merely indicates a lack of evidence to reject it.

**Can a p-value definitively prove anything?**

**No, a smaller p-value cannot definitively prove that the null hypothesis is false.  It can only provide stronger evidence against it.** Similarly, **we cannot definitively prove Fido's innocence or guilt using a p-value.**  

Here's why:

* **Hypothesis testing is about evidence, not proof.** A p-value helps us understand the likelihood of observing our data if the null hypothesis were true. A small p-value suggests that our data would be unlikely if the null hypothesis were true, leading us to reject the null hypothesis. However, it doesn't "prove" the alternative hypothesis is true.
* **There's always a chance of error.** We could obtain a small p-value by chance, even if the null hypothesis were true (a Type I error).  Conversely, we could fail to reject the null hypothesis even if it were false (a Type II error).
* **The p-value depends on the null hypothesis and its assumptions.** The p-value is calculated based on the assumption that the null hypothesis is true. If the null hypothesis is poorly defined or its assumptions are violated, the p-value may be misleading. 

**Rather than seeking definitive proof, we use p-values and hypothesis testing to make informed decisions based on the strength of the evidence.**  A smaller p-value provides stronger evidence against the null hypothesis, but it never completely rules out the possibility that the null hypothesis could be true.

### **Problem Introduction**

This analysis is inspired by the famous Fisher's Tea Experiment, where Dr. Muriel Bristol claimed she could taste whether milk was added before or after tea. Fisher, skeptical of this claim, proposed a hypothesis test. In our adaptation of this experiment, we examine whether students in STA130 can correctly identify if milk or tea was poured first in a single cup of tea. Out of 80 students, 49 correctly identified the pouring order.

### **Relationship Between This Experiment and the Original**

The key differences between the STA130 experiment and Fishers original experiment are:

1. Sample Size: Fisher's original test involved only 8 cups, while this experiment involves 80 students, each tasting one cup.
2. Population: Dr. Bristol, the subject of the original experiment, had a personal claim about her ability to distinguish the order of pouring. In this context, the STA130 students are a more general population, with no prior claim about their ability to distinguish the order.

### **Null and Alternative Hypotheses**

**Null Hypothesis (H)**
**Formal Statement:** The students are randomly guessing, i.e., the probability that a student correctly identifies whether milk or tea was poured first is 0.5.

**Mathematical Form:**

$H_0$:  = 0.5
where  is the population proportion of students who can correctly identify the pouring order.

**Informal Statement:** The students are just guessing, and their chance of being correct is 50%.

**Alternative Hypothesis (H)**
**Formal Statement:** The students are not randomly guessing; they have a better-than-chance ability to correctly identify the order of pouring.

**Mathematical Form:**
$H_1$:  > 0.5

**Informal Statement:** The students have some ability to correctly identify whether the milk or tea was poured first, better than just random guessing.

### **Quantitative Analysis**

**Observed Data**
- Sample Size (): 80 students
- Correct Responses: 49 students
- Sample Proportion ($\hat{p}$): 
$\hat{p}$ = 49/80 = 0.6125
We want to assess whether this sample proportion is significantly greater than 0.5.

**Methodology**
We will conduct a **hypothesis test for a population proportion** using the following steps:

1. **Assume H is true**, which means the probability of a correct response is 0.5.
2. **Compute the test statistic**, which follows a normal distribution under the null hypothesis.
The test statistic is given by:

z = ($\hat{p}$ - $p_0$) / $\sqrt{((p_0)(1-p_0)) / n}$

where:
- $\hat{p}$ is the sample proportion (0.6125)
- $p_0$ is the population proportion under the null hypothesis (0.5)
- n is the sample size (80)

3. **Estimate the p-value** using the standard normal distribution.
4. **Draw conclusions** based on the p-value.

```python
import numpy as np
from scipy.stats import norm

# Set seed for reproducibility
np.random.seed(130)

# Given data
n = 80  # sample size
correct_responses = 49  # number of correct responses
p_hat = correct_responses / n  # sample proportion
p_0 = 0.5  # null hypothesis proportion

# Calculate the z-score (test statistic)
z = (p_hat - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)

# Calculate the p-value
p_value = 1 - norm.cdf(z)  # One-tailed test

# Output results
z, p_value
```

This will compute the z-score and the p-value, which we will interpret to assess whether we reject or fail to reject the null hypothesis.

**Supporting Visualizations (Optional)**

We can simulate the sampling distribution of the proportion under the null hypothesis to visually compare the observed sample proportion to the distribution.

### **Findings and Discussion**

- **p-value:** The p-value indicates the probability of obtaining a sample proportion as extreme as 0.6125, assuming the null hypothesis is true. A low p-value suggests that the observed result is unlikely under the null hypothesis.
